```python
import pandas as pd
df1 = pd.read_excel("C:/Users/user1/Downloads/datasets/cust_data.xlsx", sheet_name="Sheet1")
```

```python
df1
```

```python
df2 = pd.read_excel("C:/Users/user1/Downloads/datasets/cust_data.xlsx", sheet_name="Sheet2")
df2
```

```python
# define a new row being added
new_row = { "Customer_ID":17,
            "Genre":0,
            "Age": 34,
            "Annual_Income":18,
            "Spending_Score":90         
           }

df1 = df1.append(new_row, ignore_index=True)
df1
```

```python
df2['Name'], df2['Camp']
```

```python
names = ["Tam","Ram","Sam"]
ind=[]
for i in names: ind.append(df2[df2['Name'] == i].index) 
ind
```

```python
for i in range(0, len(df2['Camp'])):
    for j in ind:
        if i == j: df2['Camp'][i] = df2['Camp'][i] + 1
        else: df2['Camp'][i] = df2['Camp'][i]
```

```python
# Initalizing a writer object to edit an excel;
# print(os.getcwd()) --> check the current working directory and specify the correct file path to "ExcelWriter"
writer = pd.ExcelWriter('cust_data.xlsx', engine='xlsxwriter')

df1.to_excel(writer, sheet_name="Sheet1", index=False)
df2.to_excel(writer, sheet_name="Sheet2", index=False)

writer.save()
```
